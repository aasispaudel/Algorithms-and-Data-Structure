```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
scale = 1.0

naive = []
bottom = []
closed = []
matrix = []
M = list(range(1, 35, 1))
N = M[:int(len(M)*scale)]

with open('/home/umbrella/aid_hw/a3/diff_file.txt') as file:
    for i in file.readlines():
        n, b, c, m = i.split(',')
        naive.append(int(n))
        bottom.append(int(b))
        closed.append(int(c))
        matrix.append(int(m))
        
plt.figure()
ax = plt.subplot(111)

ax.set_yscale('log')

ax.plot(N, naive, 'm', label='Naive recursion')
ax.plot(N, bottom, 'y', label='Bottom up')
ax.plot(N, closed, 'k', label='Closed form')
ax.plot(N, matrix, 'r', label='Matrix method')

plt.legend()
plt.title('Different fibonacci graph observation')
plt.xlabel("The nth Fibonacci")
plt.ylabel("Time taken")
plt.savefig("theCrown.pdf")
```

```python
N = []
naive = []
bottom = []
closed = []
matrix = []

with open('/home/umbrella/Downloads/data.txt') as file:
    for i in file.readlines():
        n, nav, b, c, m = i.split()
        N.append(n)
        naive.append(float(nav))
        bottom.append(float(b))
        closed.append(float(c))
        matrix.append(float(m))
        
plt.figure()
ax = plt.subplot(111)

ax.set_yscale('log')

ax.plot([N[i] for i in range(0, len(N), 2)], [naive[i] for i in range(0, len(naive), 2)], 'm', label='Naive recursion')
ax.plot([N[i] for i in range(0, len(N), 2)], [bottom[i] for i in range(0, len(bottom), 2)], 'y', label='Bottom up')
ax.plot([N[i] for i in range(0, len(N), 2)], [closed[i] for i in range(0, len(closed), 2)], 'k', label='Closed form')
ax.plot([N[i] for i in range(0, len(N), 2)], [matrix[i] for i in range(0, len(matrix), 2)], 'r', label='Matrix method')

plt.legend()
plt.title('Different fibonacci graph observation')
plt.xlabel("The nth Fibonacci")
plt.ylabel("Time taken")
# plt.show()
plt.savefig("graph.pdf")
```
